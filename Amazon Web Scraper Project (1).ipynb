### Scrap product data from website

```python
#Import liabraries
from bs4 import BeautifulSoup
import requests
import lxml
import csv
```

```python
URL = "https://www.amazon.in/Apple-Headphones-Cancellation-Transparency-Personalised/dp/B0DGJ61KZ8/ref=sr_1_3?crid=2NXMIA82PQ121&dib=eyJ2IjoiMSJ9.vzX334wLJ6V0Kd-EJcJMm5sraP85JHUPoalcqPVLtStrAFgzzQgW3oK6HA8qG1shCFxGJnHsyezAtnQjc-X8fM-ziA1RwTbMa9oZ1RC0_yPYOdB-mX0HbHpYvWxFhYSZRAF8ElohiImGXZ16wmR5jx-ktA2Xs6bslg0bM2FMqCigOQyH1ytANB-wpJD49oIIA5-6wIF7HxWJTvykOqCi4axub7rvSoJt01Zr-LLjOuY.xReW4YY1pbdbi4Sec_Cy6Idec4QjhMa1cF2jR952Xzg&dib_tag=se&keywords=apple%2Bairpods%2Bpro%2Bmax&nsdOptOutParam=true&qid=1744254152&sprefix=apple%2Bairpods%2Bpro%2Bma%2Caps%2C289&sr=8-3&th=1"

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36 Edg/134.0.0.0"}

response = requests.get(URL, headers=headers)

response.status_code ==200

if response.status_code ==200:
    #print(response.status_code)
    html_content = response.text
else:
    print("fetching error", response.status_code)
```

```python
soup = BeautifulSoup(html_content, 'lxml')
print(soup.prettify())
```

```python
product_title = soup.find("span", id="productTitle").text.strip()
print(product_title)
```

```python
product_price = soup.find("span", class_="a-price-whole").text.strip()
print(product_price)
```

```python
product_rating = soup.find("span", id="acrPopover" ).text.strip()
print(product_rating)
```

```python
product_bp = soup.find("ul", class_="a-unordered-list a-vertical a-spacing-mini").text.strip()
print(product_bp)
```

```python
product_description = soup.find("div", id="productDescription").text.strip()
print(product_description)
```

```python
product_insidebox = soup.find("div", id="whatsInTheBoxDeck").text.strip()
print(product_insidebox)
```

```python
# Save to Downloads folder (or any safe path)

file_path = (r"C:\Users\biswa\OneDrive\Documents\Amazon Web Scarp.csv")

with open(file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow([
        "product_title", "product_price", "product_rating",
        "product_bp", "product_description", "product_insidebox"
    ])

    writer.writerow([
        product_title, product_price, product_rating,
        product_bp, product_description, 
        product_insidebox
    ])

print("✅ Data written successfully!")
```
